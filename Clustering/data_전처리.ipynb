```python
import pandas as pd
```

```python
# CSV 파일 경로 설정
csv_files = ['./SVIL_DATA_20230521/youtube_boy_heavy/20230518_youtube_boy_heavy.csv', 
             './SVIL_DATA_20230521/youtube_boy_light/20230518_youtube_boy_light.csv', 
             './SVIL_DATA_20230521/youtube_girl_heavy/20230518_youtube_girl_heavy.csv', 
             './SVIL_DATA_20230521/youtube_girl_light/20230518_youtube_girl_light.csv']
```

```python
# 빈 DataFrame 생성
combined_df = pd.DataFrame()
```

```python
# CSV 파일 읽어오기 및 DataFrame에 추가
for file in csv_files:
    df = pd.read_csv(file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
# 결과 확인
combined_df
```

```python
# 기존 DataFrame에서 'comment' 열에 '구독자'라는 단어가 있는 행을 추출할 DataFrame 생성
subscribers_df = combined_df[combined_df['comment'].str.contains('구독자')]
```

```python
# 'view'와 'comment' 열만 추출
subscribers_view_comment_df = subscribers_df[['view', 'comment']]
```

```python
# 'view' 열의 숫자 형식 변경
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace('조회수', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace('회', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace(',', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].astype(float)
```

```python
# 'comment' 열의 데이터 처리
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].str.replace('구독자', '')
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].str.replace('만명', '')
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].astype(float)*10000
```

```python
subscribers_view_comment_df
```

```python
subscribers_df
```

```python
# 기존 DataFrame에서 'comment' 열에 '구독자'라는 단어가 없는 행을 추출할 DataFrame 생성
non_subscribers_df = combined_df[~combined_df['comment'].str.contains('구독자')]
```

```python
# 'tag' 값이 0인 행만 선택
non_subscribers_df = non_subscribers_df[non_subscribers_df['tag'] == 0]
```

```python
non_subscribers_df
```

```python
# 'view'와 'comment' 열만 추출
non_subscribers_view_comment_df = non_subscribers_df[['view', 'comment']]
```

```python
# 'view' 열의 숫자 형식 변경
non_subscribers_view_comment_df['view'] = non_subscribers_view_comment_df['view'].astype(float)
```

```python
# 'comment' 열의 데이터 처리
non_subscribers_view_comment_df['comment'] = non_subscribers_view_comment_df['comment'].str.replace('구독자', '')
non_subscribers_view_comment_df['comment'] = non_subscribers_view_comment_df['comment'].str.replace('만명', '')
non_subscribers_view_comment_df['comment'] = non_subscribers_view_comment_df['comment'].astype(float)*10000
```

```python
non_subscribers_view_comment_df
```

```python
# 구독자 DataFrame을 CSV 파일로 저장
subscribers_view_comment_df.to_csv('subscribers.csv', index=False)
# 비구독자 DataFrame을 CSV 파일로 저장
non_subscribers_view_comment_df.to_csv('non_subscribers.csv', index=False)
```
