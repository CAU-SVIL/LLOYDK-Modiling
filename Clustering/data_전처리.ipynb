```python
import pandas as pd
```

```python
# CSV íŒŒì¼ ê²½ë¡œ ì„¤ì •
csv_files = ['./SVIL_DATA_20230521/youtube_boy_heavy/20230518_youtube_boy_heavy.csv', 
             './SVIL_DATA_20230521/youtube_boy_light/20230518_youtube_boy_light.csv', 
             './SVIL_DATA_20230521/youtube_girl_heavy/20230518_youtube_girl_heavy.csv', 
             './SVIL_DATA_20230521/youtube_girl_light/20230518_youtube_girl_light.csv']
```

```python
# ë¹ˆ DataFrame ìƒì„±
combined_df = pd.DataFrame()
```

```python
# CSV íŒŒì¼ ì½ì–´ì˜¤ê¸° ë° DataFrameì— ì¶”ê°€
for file in csv_files:
    df = pd.read_csv(file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
# ê²°ê³¼ í™•ì¸
combined_df
```

```python
# ê¸°ì¡´ DataFrameì—ì„œ 'comment' ì—´ì— 'êµ¬ë…ì'ë¼ëŠ” ë‹¨ì–´ê°€ ìˆëŠ” í–‰ì„ ì¶”ì¶œí•  DataFrame ìƒì„±
subscribers_df = combined_df[combined_df['comment'].str.contains('êµ¬ë…ì')]
```

```python
# 'view'ì™€ 'comment' ì—´ë§Œ ì¶”ì¶œ
subscribers_view_comment_df = subscribers_df[['view', 'comment']]
```

```python
# 'view' ì—´ì˜ ìˆ«ì í˜•ì‹ ë³€ê²½
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace('ì¡°íšŒìˆ˜', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace('íšŒ', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].str.replace(',', '')
subscribers_view_comment_df['view'] = subscribers_view_comment_df['view'].astype(float)
```

```python
# 'comment' ì—´ì˜ ë°ì´í„° ì²˜ë¦¬
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].str.replace('êµ¬ë…ì', '')
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].str.replace('ë§Œëª…', '')
subscribers_view_comment_df['comment'] = subscribers_view_comment_df['comment'].astype(float)*10000
```

```python
# ê¸°ì¡´ DataFrameì—ì„œ 'comment' ì—´ì— 'êµ¬ë…ì'ë¼ëŠ” ë‹¨ì–´ê°€ ì—†ëŠ” í–‰ì„ ì¶”ì¶œí•  DataFrame ìƒì„±
non_subscribers_df = combined_df[~combined_df['comment'].str.contains('êµ¬ë…ì')]
```
